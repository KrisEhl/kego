```python
def assert_equal_list_dicts(
    list_1: list[dict],
    list_2: list[dict],
    name_list_1: str = "list_1",
    name_list_2: str = "list_2",
):
    def normalize(d):
        return tuple(sorted(d.items()))

    from collections import Counter

    counter_1 = Counter([normalize(d) for d in list_1])
    counter_2 = Counter([normalize(d) for d in list_2])

    if counter_1 != counter_2:
        missing_in_1 = list((counter_2 - counter_1).elements())
        missing_in_2 = list((counter_1 - counter_2).elements())
        raise AssertionError(
            f"{name_list_1} != {name_list_2}\n"
            f"Missing in {name_list_1}: {missing_in_1}\n"
            f"Missing in {name_list_2}: {missing_in_2}"
        )
```

```python
x = [{"2": 2, "awda": [1, 2, 3]}, {"a": 2, "13": 13}]
y = [{"2": 2, "awda": [1, 2, 3]}, {"a": 2}]

assert_equal_list_dicts(x, y)
```
